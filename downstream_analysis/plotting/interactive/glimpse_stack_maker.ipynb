# If there are corrupt glimpses

```python
from macrohet import visualise
import re
```

```python
import pickle
```

```python
base_dir = '/mnt/DATA/macrohet/'
```

```python
# Specify the path to your pickle file
sc_df_path = os.path.join(base_dir, "results/preliminary_sc_measures/sc_dfs/sc_df_40_interp.pkl" )
#"/mnt/DATA/macrohet/results/preliminary_sc_measures/sc_dfs/sc_df_>40.pkl"

# Load the pickle file
with open(sc_df_path, "rb") as file:
    df = pickle.load(file)
```

```python
from skimage.io import imshow, imsave
```

```python
glimpse_stack = list()
for fn in tqdm(glimpse_fns, total=len(glimpse_fns)):
    try:
        glimpse = imread(fn)
    except:
        print(f'Glimpse {fn} could not be loaded, regenerating glimpse')
        unique_ID = fn.split('_')[-2]
        time = int(re.search(r'\d+', fn.split('_')[-1]).group())
        glimpse = visualise.tn_glimpse_maker(unique_ID, df, time)
        imsave(fn, glimpse)
    glimpse_stack.append(glimpse)
```

```python
import numpy as np
from skimage.transform import resize

# Calculate the maximum dimensions of the images
max_height = max(image.shape[0] for image in glimpse_stack)
max_width = max(image.shape[1] for image in glimpse_stack)
num_channels = glimpse_stack[0].shape[2]  # Assuming all images have the same number of channels

# Create an empty array to store the scaled images
scaled_images = np.zeros((len(glimpse_stack), max_height, max_width, num_channels), dtype=np.uint8)

# Scale each image to the maximum dimensions
for i, image in tqdm(enumerate(glimpse_stack), total=len(glimpse_stack)):
    scaled_images[i] = resize(image, (max_height, max_width), preserve_range=True)

# The scaled images are now stored in the 'scaled_images' array
```

```python
import napari
```

```python
v = napari.Viewer()
v.add_image(scaled_images)
```
