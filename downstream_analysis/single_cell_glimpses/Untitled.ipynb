```python
import cv2

def apply_gaussian_blur(image_path, output_path, kernel_size, sigma):
    # Load the image
    image = cv2.imread(image_path)
    
    # Apply Gaussian blur
    blurred_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)
    
    # # Save the output image
    # cv2.imwrite(output_path, blurred_image)
    return blurred_image

# Specify the input image path, output image path, kernel size, and sigma
input_image_path = 'input.jpg'
output_image_path = 'blurred_output.jpg'
kernel_size = 5  # Should be an odd number
sigma = 1.5

# Apply Gaussian blur to the input image and save the result
apply_gaussian_blur(input_image_path, output_image_path, kernel_size, sigma)
```
