```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create a figure
fig, ax = plt.subplots(figsize=(12, 8))

# Function to draw a box with text
def draw_box(ax, text, x, y, width=2, height=1):
    rect = mpatches.Rectangle((x, y), width, height, fill=True, edgecolor="black", facecolor="lightgray")
    ax.add_patch(rect)
    ax.text(x + width / 2, y + height / 2, text, ha="center", va="center", fontsize=12)

# Draw boxes
draw_box(ax, "Human iPSC Culture", 0, 7)
draw_box(ax, "Monocyte Factories Setup", 0, 5)
draw_box(ax, "Monocyte Differentiation", 0, 3)
draw_box(ax, "Macrophage Preparation", 0, 1)
draw_box(ax, "Infection with iPSDMs", 0, -1)

# Draw arrows
ax.arrow(1, 6.5, 0, -0.5, head_width=0.1, head_length=0.2, fc="black", ec="black")
ax.arrow(1, 4.5, 0, -0.5, head_width=0.1, head_length=0.2, fc="black", ec="black")
ax.arrow(1, 2.5, 0, -0.5, head_width=0.1, head_length=0.2, fc="black", ec="black")
ax.arrow(1, 0.5, 0, -0.5, head_width=0.1, head_length=0.2, fc="black", ec="black")

# Set axis limits and hide axes
ax.set_xlim(-1, 3)
ax.set_ylim(-2, 8)
ax.axis('off')

plt.title("Methodology for Human iPSC Culture, Monocyte Differentiation, and Macrophage Infection", fontsize=14)
plt.show()
```
