# Inspect h5

```python
# Open the HDF5 file for reading
import h5py

# Open the HDF5 file for reading
with h5py.File(os.path.join(base_dir, f'labels/relocalisation/relocalised_{row, column}.h5'), 'r') as file:

    # List all the groups in the file
    print('Groups:')
    for group in file.keys():
        print(group)

    # Get the first group
    group = list(file.keys())[0]

    # List all the datasets in the group
    print('\nDatasets in %s:' % group)
    for dataset in file[group].keys():
        print(dataset)

    # Get the first dataset
    dataset = list(file[group].keys())[0]

    # Access the data in the dataset
    data = file[group][dataset][()]
    print('\nData in %s/%s:' % (group, dataset))
    print(data)

```
