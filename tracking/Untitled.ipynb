```python
from btrack import datasets
```

```python
segmentation = datasets.example_segmentation()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(segmentation[-1][500:1200, 900:1400].T)
```

```python
from skimage.io import imsave
```

```python
imsave('/home/dayn/Pictures/thesis/semantic.png', segmentation[-1][500:1200, 900:1400].T)
```

```python
plt.imshow((segmentation[-1][500:1200, 900:1400].T))
plt.axis('off')
plt.savefig('/home/dayn/Pictures/thesis/semantic.png', bbox_inches = 'tight')
```

```python
from skimage.morphology import label
```

```python
plt.imshow(label(segmentation[-1][500:1200, 900:1400].T))
plt.axis('off')
plt.savefig('/home/dayn/Pictures/thesis/instance.png', bbox_inches = 'tight')
```

```python
imsave('/home/dayn/Pictures/thesis/instance.png', label(segmentation[-1][500:1200, 900:1400].T))
```

```python
fig = plt.figure()
ax = plt.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)
ax.imshow(label(segmentation[-1][500:1200, 900:1400].T))
plt.savefig('/home/dayn/Pictures/thesis/instance.png', bbox_inches = 'tight', dpi = 300)
```
